Configura el eje x del gráfico para evitar la notación científica, crea un diagrama de caja para la columna `'poblacion'` del DataFrame `data`, añade una línea vertical en la media de `'poblacion'` (en color amarillo), establece el título del gráfico como 'Casos confirmados de COVID' y muestra el gráfico.

# Grafico de histograma

Importa las bibliotecas `numpy` como `np` y `seaborn` como `sns`. Luego, simula un conjunto de datos con una distribución normal, crea un histograma con una curva de densidad sobrepuesta usando `seaborn`, y muestra el gráfico. El histograma se puede crear de dos maneras: directamente con los datos (`x`) o usando un DataFrame (`data`).

```python
import numpy as np
import seaborn as sns

# Simulación de datos
rng = np.random.RandomState(1)
x = rng.normal(0, 1, size = 1000)
data = {'x': x}

# Histograma
sns.histplot(x = x, kde = True)

# Equivalente a:
sns.histplot(x = "x", kde = True, data = data)
ptl.show()
```

## Grafica Pastel

Importa la biblioteca `matplotlib.pyplot` como `plt`. Luego, crea un gráfico de pastel para mostrar la distribución de casos confirmados en diferentes estados, con etiquetas, porcentajes y colores personalizados. Ajusta el tamaño de la figura, añade un título y muestra el gráfico.

```python
import matplotlib.pyplot as plt

# Data extracted from the image
states = ['AGUASCALIENTES', 'BAJA CALIFORNIA', 'BAJA CALIFORNIA SUR', 'CAMPECHE', 'CHIAPAS']
cases = [2, 23, 6, 8, 6]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(cases, labels=states, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])
plt.title('Casos confirmados 15-06-2023')
plt.show()
```
